```python
import pandas as pd
import numpy as np
import requests
from io import BytesIO
import zipfile
import os
from scipy.stats import chi2_contingency

# Load the tabulatedVotes.csv file
data = pd.read_csv("tabulatedVotes.csv")
data.head()
```

```python
def fleiss_kappa(data):
    n_subjects, n_categories = data.shape
    n_raters = data.sum(axis=1)  # Maximum number of raters for any subject    
    N = n_raters[0] * n_subjects
    p = np.sum(data, axis=0) / N
    P_i = (1/(n_raters * (n_raters - 1))) * (np.sum(data ** 2, axis=1) - n_raters)
    P_bar = np.mean(P_i)
    P_e_bar = np.sum(p ** 2)
    kappa = (P_bar - P_e_bar) / (1 - P_e_bar)
    return kappa
```

```python
selected_columns = ['A', 'D', 'F', 'S', 'H', 'N' , 'S']
filtered_data = data[selected_columns].to_numpy()
kappa = fleiss_kappa(filtered_data)
print("Fleiss' Kappa:", kappa)
print("Fleiss Kappa value indicates that there is moderate inter annotation agreement")
```

```python
# Loading the annotation data
annotated_data = pd.read_csv('emotion_annotation.csv')
real_annotation = pd.read_csv('real_annotation.csv')

# Merging the two datasets on the 'filename' column
merged_data = pd.merge(real_annotation, annotated_data, on='filename')
# Creating a contingency table
contingency_table = pd.crosstab(merged_data['HAPPY_x'], merged_data['HAPPY_y'])

# Displaying the contingency table
contingency_table
```

```python
chi2, p, _, _ = chi2_contingency(contingency_table)
print("Chi-squared statistic:", chi2)
print("P-value:", p)

threshold = 0.05
if p < threshold:
    print("Rejected null hypothesis: There is a significant association between 'Happy' and 'Sad' flags.")
else:
    print("Fail to reject null hypothesis: There is no significant association between 'Happy' and 'Sad' flags.")
```
